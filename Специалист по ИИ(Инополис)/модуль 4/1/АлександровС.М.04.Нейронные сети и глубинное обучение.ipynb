```python
def convolution(a, b):        # пропишем функцию свертки, она соответствуте описанию, которое приведено выше
    sum = 0
    for i in range(len(a)):
        for j in range(len(a[0])):
            sum += a[i][j] * b[i][j]
    return sum
```
