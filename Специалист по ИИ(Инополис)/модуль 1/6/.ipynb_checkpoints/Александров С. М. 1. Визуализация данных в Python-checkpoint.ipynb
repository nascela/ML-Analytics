```python
import pandas as pd
import matplotlib # библиотека для построения графиков
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Bike.csv')
df.head()
```

```python
df[['datetime','season']].groupby('season').count().plot(kind='bar', legend=False)
plt.title('Распределение сезонов')
plt.xlabel('сезон')
plt.ylabel('время')
plt.show()
```

```python
df[['datetime','temp']].groupby('temp').count().plot(kind='bar', legend=False)
plt.title('Распределение сезонов')
plt.xlabel('температура')
plt.ylabel('время')
plt.show()
```

```python
plt.hist(df['windspeed'],bins =20, edgecolor = 'pink')
plt.title('Распределение по скорости ветра')
plt.xlabel('сила ветра')
plt.ylabel('Кол-во')
plt.show()
```

```python
sns.distplot(df['windspeed'])
plt.title('Распределение по скорости ветра')
plt.xlabel('Скорость ветра')
plt.ylabel('Кол-во')
plt.show()
```

```python
spring = df[(df['season']==1)]
summer = df[(df['season']==2)]
autumn = df[(df['season']==3)]
winter = df[(df['season']==4)]
plt.hist([spring['season'], summer['season'],autumn['season'],winter['season']],color=[ "b", "r","y","g"], edgecolor = 'pink', label=['весна','лето','осень','зима' ])
plt.legend()
plt.xlabel('season')
plt.ylabel('Кол-во  сезонов')
plt.title('Распределение сезонов в зависимости от количества')
plt.show()
```
